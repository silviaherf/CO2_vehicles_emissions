### We open the second dataset about GHG emissions worldwide

```python
# we open the second dataset about GHG emissions worldwide
emissions2 = man.open_cvs()
emissions2.head()
##Greenhouse_Gas_Emissions_worldwide is the file in input folder
```

```python
#we will rename columns removing blank spaces to avoid  problems in future by calling a function from the
# file columns.py
clean.rename_columns(emissions2)
emissions2.head()
```

```python
#we will analyze biggest areas/countries emissions data
biggest_emissors=emissions2[['Country_or_Area','Year','co2_gigagrams']]
biggest_emissors.columns=['Country_or_Area','Year','CO2_Tonnes']
biggest_emissors=biggest_emissors.groupby('Country_or_Area').agg({'CO2_Tonnes':['sum','mean']})
biggest_emissors.head()
```

```python
biggest_emissors=biggest_emissors.sort_values(by=[('CO2_Tonnes','sum')],ascending=False).head(10)
biggest_emissors
```

```python
#I prefer to manage Tonnes in this case
biggest_emissors=biggest_emissors.apply(lambda x:(x/1e6),axis=1)
clean.rounding(biggest_emissors)
biggest_emissors=biggest_emissors.reset_index()
biggest_emissors
```
